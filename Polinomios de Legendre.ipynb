```
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff, factorial, simplify, legendre
```

```
# Variables simb√≥licas
x = symbols('x')
n_values = range(6)  # n = 0, 1, ..., 5
```

Define x como una variable simb√≥lica para manipular polinomios.

Define un rango de valores para ùëõ (de 0 a 5) para los distintos grados de los polinomios

```
# Inciso (a): Graficar los polinomios de Legendre Pn(x) en un solo panel
x_vals = np.linspace(-1, 1, 400)
plt.figure(figsize=(8, 6))

for n in n_values:
    Pn = legendre(n, x)
    Pn_vals = [Pn.evalf(subs={x: val}) for val in x_vals]
    plt.plot(x_vals, Pn_vals, label=f'$P_{n}(x)$')

plt.title('Polinomios de Legendre $P_n(x)$ para $n=0,\\dots,5$')
plt.xlabel('$x$')
plt.ylabel('$P_n(x)$')
plt.legend()
plt.grid(True)
plt.show()
```

Define los puntos ùë• en el intervalo [‚àí1,1].

Calcula ùëÉ_ùëõ(ùë•)usando la funci√≥n legendre.

```
# Inciso (b): Verificaci√≥n gr√°fica para n=3
n = 3
P3 = legendre(3, x)
P4 = legendre(4, x)

# Derivada de P3(x)
P3_prime = diff(P3, x)

# Relaci√≥n a graficar
lhs = (1 - x**2) * P3_prime
rhs = (n + 1) * x * P3 - (n + 1) * P4

# Evaluaci√≥n num√©rica para la gr√°fica
lhs_vals = [lhs.evalf(subs={x: val}) for val in x_vals]
rhs_vals = [rhs.evalf(subs={x: val}) for val in x_vals]

plt.figure(figsize=(8, 6))
plt.plot(x_vals, lhs_vals, label='$LHS: (1 - x^2) P_3\'(x)$', linestyle='--')
plt.plot(x_vals, rhs_vals, label='$RHS: 4x P_3(x) - 4 P_4(x)$', linestyle='-.')
plt.title('Relaci√≥n gr√°fica para $n=3$')
plt.xlabel('$x$')
plt.ylabel('Valor')
plt.legend()
plt.grid(True)
plt.show()
```

```
# Inciso (c): Verificaci√≥n algebraica de la relaci√≥n para n=3
algebraic_verification = simplify(lhs - rhs)
print("Verificaci√≥n algebraica (c):", algebraic_verification)
```

```
# Inciso (e): Verificaci√≥n de la f√≥rmula de Rodrigues para n=0,1,...,5
for n in n_values:
    rodrigues = (1 / (2**n * factorial(n))) * diff((x**2 - 1)**n, x, n)
    rodrigues_check = simplify(rodrigues - legendre(n, x))
    print(f"Verificaci√≥n de Rodrigues para n={n}:", rodrigues_check)
```
