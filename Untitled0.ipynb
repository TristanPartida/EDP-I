```python
# prompt: graficar y=x^2

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 400)
y = x**2

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y = x^2")
plt.title("Gráfica de y = x^2", color='green')
plt.grid(True)
plt.show()
```
