### Graficas 3D

```python
from sympy.plotting import plot3d #Llamamos a la funcion plot3d la cual nos ayuda a graficar funciones x,y
```

```python
x,y=symbols('x y') #
```

```python
plot3d(x*y)
```

```python
plot3d(x*y, (x,-5,5),(y,-5,5))
```

```python
plot3d(x*y, -x*y, (x,-5,5),(y,-5,5))
```

```python
plot3d((x**2 + y**2, (x,-5,5),(y,-5,5)),(x*y,(x,-3,3),(y,-3,3)))
```

```python
from sympy.plotting import plot3d_parametric_line
```

```python
plot3d_parametric_line(cos(s),sin(s),s)
```

```python
plot3d_parametric_line((cos(s),sin(s),s,(s,-5,5)),(sin(s),s**2,s,(s,-5,5)))
```

```python
from sympy.plotting import plot3d_parametric_surface
```

```python
u, v =symbols('u v')
```

```python
plot3d_parametric_surface(cos(u+v),sin(u-v),u-v)
```

```python
plot3d_parametric_surface(cos(v)*cos(u),cos(v)*sin(u),sin(v))
```

Funciones impl√≠citas

```python
from sympy import plot_implicit, symbols, Eq, And
```

En SymPy, la funci√≥n plot_implicit() se utiliza para graficar ecuaciones impl√≠citas. Las ecuaciones impl√≠citas son aquellas que no se expresan de forma expl√≠cita, es decir, no se resuelven para una sola variable. Esto es √∫til cuando quieres graficar curvas o superficies definidas por ecuaciones que involucran m√°s de una variable.
plot_implicit(): Permite graficar ecuaciones impl√≠citas en 2D o 3D.
symbols(): Se usa para definir las variables simb√≥licas (por ejemplo,

Eq(): Es una funci√≥n que define una ecuaci√≥n simb√≥lica en Python.
And(): Permite combinar varias condiciones l√≥gicas, como en el caso de que quieras graficar m√∫ltiples ecuaciones simult√°neamente.

```python
x, y = symbols(' x y')
```

Se utiliza para definir las variables simb√≥licas que vas a usar en tus expresiones matem√°ticas y ecuaciones simb√≥licas.

```python
p1 = plot_implicit(Eq(x**2 + y**2, 5)) #Se utiliza para graficar una ecuaci√≥n impl√≠cita en SymPy.#
```
